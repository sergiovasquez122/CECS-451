```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def utility(start, middle, terminal, gamma, length):
    utility = 0.0
    t = 1.0
    discount = gamma
    utility += start
    t += 1
    for _ in range(length - 2):
        discount = gamma ** t
        utility += middle * discount
        t += 1
    utility += terminal * discount
    return utility

def eval_choices(g):
    gamma = g
    print("gamma:", gamma)
    up = utility(50.0, -1.0, 10.0, gamma, 101)
    down = utility(-50.0, 1.0, -10.0, gamma, 101)
    print("up:", up)
    print("down:", down)
    print("The best choice is:", "up" if (up > down) else "down")
    
arr = np.linspace(0, 1, 101)
up = np.zeros(101)
down = np.zeros(101)

#print(up, down)

for i in range(arr.size):
    up[i] = utility(50.0, -1.0, 10.0, arr[i], 101)
    down[i] = utility(-50.0, 1.0, -10.0, arr[i], 101)
```

```python
plt.plot(arr, up)
plt.rcParams["figure.figsize"] = (5,5)
plt.plot(arr, down)
plt.legend
```
