```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
X = cancer.data
y = cancer.target
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=2)
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
g = plot_tree(clf)
```

```python
from sklearn.ensemble import BaggingClassifier
estimators = []
scores = []
for i in range(10, 40):
    clf = BaggingClassifier(n_estimators=i).fit(X_train, y_train)
    estimators.append(i)
    scores.append(accuracy_score(y_test, clf.predict(X_test)))
```

```python
plt.plot(estimators, scores)
plt.xlabel('n_estimators')
plt.ylabel('bagging score')
```

```python
from sklearn.ensemble import AdaBoostClassifier
estimators = []
scores = []
for i in range(10, 40):
    clf = BaggingClassifier(n_estimators=i).fit(X_train, y_train)
    estimators.append(i)
    scores.append(accuracy_score(y_test, clf.predict(X_test)))
```

```python
plt.plot(estimators, scores)
plt.xlabel('n_estimators')
plt.ylabel('boost score')
```

```python
from sklearn.ensemble import RandomForestClassifier
estimators = []
features = []
scores = []
for i in range(10, 50):
    for j in range(1, len(cancer.feature_names)):
        clf = RandomForestClassifier(n_estimators=i, max_features=j).fit(X_train, y_train)
        estimators.append(i)
        features.append(j)
        scores.append(accuracy_score(y_test, clf.predict(X_test)))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (20, 20))
ax = plt.axes(projection ="3d")
 
my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(estimators, features, scores,
                    alpha = 0.8,
                    c = (estimators),
                    cmap = my_cmap)

plt.title("Random Forest Plot")
ax.set_xlabel('n_estimators', fontweight ='bold')
ax.set_ylabel('max_features', fontweight ='bold')
ax.set_zlabel('forest_score', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
```
